```python
from network import scGen
import scanpy.api as sc
import numpy as np
import wget
```

```python
path_to_save = "../results"
sc.settings.figdir = path_to_save
```

Downloading the train and validation data for Hagai et al. dataset and saving them in data folder

```python
train_url = "https://www.dropbox.com/s/enwuvxb47dpfv0k/train_all_lps6.h5ad?dl=1"
validation_url = "https://drive.google.com/uc?export=download&id=1dHr6jlKWvBGVpuT7Nyee3vBMm2W3VmLA"
t = wget.download(train_url, "../data/train_LPS6.h5ad")
v = wget.download(validation_url, "../data/valid_LPS6.h5ad")
```


Reading train and validation data

```python
train =  sc.read("../data/train_LPS6.h5ad")
validation = sc.read("../data/valid_LPS6.h5ad")
train.obs["cell_type"] = train.obs["species"]
validation.obs["cell_type"] = validation.obs["species"]
#training cells- here we treat species as they are cell types 
c_in = ['mouse', 'pig', 'rabbit',]
#heldout cells (only stimulated ones)
c_out = [ 'rat']
```

Lets now make scGen object with training, validation and condition labels with training and held out cells

```python
scg = scGen(train,validation,{"ctrl":"unst", "stim":"LPS6"}, c_in, c_out, model_path="../models/cross_species")
```

We train the model for 200 epochs 

```python
scg.train(200, early_stop_limit=15, threshold=.005)
```

Making the predictin for rat LPS6 cells




```python
rabbit = train[train.obs["species"] == "rabbit"]
rabbit_cd = rabbit[rabbit.obs["condition"] == "unst"]
rabbit_stim = rabbit[rabbit.obs["condition"] == "LPS6"]
pig = train[train.obs["species"] == "pig"]
pig_cd = pig[pig.obs["condition"] == "unst"]
pig_stim = pig[pig.obs["condition"] == "LPS6"]
mouse = train[train.obs["species"] == "mouse"]
mouse_cd = mouse[mouse.obs["condition"] == "unst"]
mouse_stim = mouse[mouse.obs["condition"] == "LPS6"]
rat = train[train.obs["species"] == "rat"]
rat_cd = rat[rat.obs["condition"] == "unst"]
rat_stim = rat[rat.obs["condition"] == "LPS6"]
```

```python
mouse_lps6_latent = scg._to_latent(mouse_stim.X.A)
mouse_cd_latent = scg._to_latent(mouse_cd.X.A)
rat_cd_latent =  scg._to_latent(rat_cd.X.A)
```

```python
min_pop = np.min([mouse_lps6_latent.shape[0],mouse_cd_latent.shape[0]])
idx  = np.random.choice(np.arange(min_pop),min_pop,replace=False)
```

Two delta vector, delta_1 which accounts for stimulation effect and delta_2 which accounts for species effect

```python
delta_1 = np.average(mouse_lps6_latent[idx],axis=0) -  np.average(mouse_cd_latent[idx],axis=0)
delta_2 = np.average(rat_cd_latent[idx],axis=0) - np.average(mouse_cd_latent[idx],axis=0)
```

Here we predict cells rat LPS6 using stimulated mouse and control rat and both delta vectors

```python
shapes = [mouse_lps6_latent.shape[0],rat_cd_latent.shape[0]]
arg_min = np.argmin(shapes)
subsample_index =  np.random.choice(np.arange(shapes[arg_min]),shapes[arg_min],replace=False)
```

```python
pred_rat_latent_1 = 0.5*(rat_cd_latent[subsample_index] +delta_1 +delta_2 + mouse_lps6_latent[subsample_index])
pred_rat1 = scg._reconstruct(pred_rat_latent_1, use_data=True)
```

```python
sc.tl.rank_genes_groups(rat, groupby="condition")
gene_list = rat.uns['rank_genes_groups']['names']["LPS6"].tolist()[0:5]
```

```python
all_Data = sc.AnnData(np.concatenate([rat_cd.X.A, rat_stim.X.A, pred_rat1]))
all_Data.obs["condition"] = ["ctrl"] * len(rat_cd) + ["real_stim"] * len(rat_stim) + \
                          ["pred_stim"] * len( pred_rat1)
all_Data.var_names = train.var_names
```

```python
scg.dr.reg_overlap_plot_cross(all_Data,gene_list,
                       path_to_save, "cross_species_first")
```

```python
all_stim = sc.AnnData(np.concatenate([rat_stim.X.A,pred_rat1
                                      ,rat_cd.X.A,rabbit_cd.X.A,
                                      rabbit_stim.X.A,pig_cd.X.A,pig_stim.X.A,
                                     mouse_cd.X.A,mouse_stim.X.A]))
all_stim.var_names = train.var_names
all_stim.obs["condition"] = ["rat_LPS6"]*len(rat_stim) + ["rat_LPS6_pred"]*len(pred_rat1)\
+["rat_ctrl"]*len(rat_cd) + ["rabbit_ctrl"]*len(rabbit_cd)  + ["rabbit_LPS6"] *len(rabbit_stim)\
+ ["pig_ctrl"]*len(pig_cd) + ["pig_LPS6"]*len(pig_stim) + ["mouse_ctrl"]*len(mouse_cd)\
+ ["mouse_LPS6"]*len(mouse_stim)
```

```python
gene_list_dot = ['Ccl3',
 'Ccl4',
 'Il1a',
 'Sod2',
 'Isg15',
 'Nfkbia',
 'Ccl5',
 'Oasl1',
 'Ier3',
 'Srgn',
 'Sdc4',
 'Slc7a2',
 'Snx10',
 'Ehd1',
 'Nampt',
 'Mmp12',
 'Usp18',
 'Cd14',
 'Acsl1',
 'Il18bp',
 'Anxa1',
 'Cxcl5',
 'Cd44',
 'Tnfaip6']
```
